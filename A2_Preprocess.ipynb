```python
import pandas as pd
```

```python
dataset = pd.read_csv("a2_dataset.csv")
```

```python
dataset.head()
```

```python
dataset = dataset[["COURSENAME","TITLE","RESOURCE_TYPE","SUBTITLE"]]
dataset.head()
```

```python
dataset['combined_title']=dataset.apply(lambda x:'%s_%s' % (x['TITLE'],x['SUBTITLE']) if (x['TITLE'] != x['SUBTITLE']) else x['TITLE'],axis=1)
dataset = dataset[['COURSENAME','combined_title','RESOURCE_TYPE']]
dataset.head()
```

```python
dataset['COURSENAME'].replace('"','',inplace=True)
dataset['combined_title'].replace('"','',inplace=True)
dataset['COURSENAME'] = dataset['COURSENAME'].str.lower()
dataset['combined_title'] = dataset['combined_title'].str.lower()
dataset['RESOURCE_TYPE'] = dataset['RESOURCE_TYPE'].str.lower()
len(dataset)
```

```python
unique_courses = dataset['COURSENAME'].unique()
unique_title = dataset.combined_title.unique()
print("Total unique courses:" + str(len(unique_courses)))
print("Total unique titles:" + str(len(unique_title)))
```

```python
for x in unique_courses:
    print(x)
```

```python
subset1 = dataset.groupby(["combined_title"]).count().sort_values(["COURSENAME"], ascending=False).reset_index()[['combined_title','COURSENAME']]
subset1
```

```python
ss1 = list(set(subset1[subset1['COURSENAME']>2]['combined_title']))
ss1
```

```python
subset_based_on_title = dataset[dataset['combined_title'].isin(ss1)]
print(len(subset_based_on_title))
print("Original: "+str(len(dataset)))
```

```python
subset_based_on_title.head()
```

```python
print(len(subset_based_on_title['COURSENAME'].unique()))
print(len(subset_based_on_title['combined_title'].unique()))
```

```python
subset2 = dataset.groupby(["COURSENAME"]).count().sort_values(["combined_title"], ascending=False).reset_index()[['COURSENAME','combined_title']]
subset2
```

```python
ss2 = list(set(subset2[subset2['combined_title']>5]['COURSENAME']))
ss2
```

```python
subset_based_on_course = dataset[dataset['COURSENAME'].isin(ss2)]
print(len(subset_based_on_course))
print("Original: "+str(len(dataset)))
```

```python
subset_based_on_course.head()
```

```python
print(len(subset_based_on_course['COURSENAME'].unique()))
print(len(subset_based_on_course['combined_title'].unique()))
```

```python
import shutil
import tempfile
import urllib.request
from urllib.parse import quote 
#urlencode, quote_plus

for title in unique_title[1:5]:
    to_fetch = "https://www.googleapis.com/books/v1/volumes?q="+quote(title)
    print(to_fetch)
    with urllib.request.urlopen(to_fetch) as response:
        html = response.read()
        #print(html)
        html_file = open(str(title)+"_google.html","w")
        html_file.write(str(html))
        html_file.close()
```
